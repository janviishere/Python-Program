```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
iris_data = pd.read_csv("/content/drive/MyDrive/Iris (1).csv")
```

```python
iris_data.head()
```

Line Plot

```python
plt.plot(iris_data["Id"], iris_data["PetalLengthCm"])
plt.show()
```

```python
plt.figure(figsize =(6,4)) #size
plt.plot(iris_data["Id"], iris_data["PetalLengthCm"])
plt.xlabel("Id")
plt.ylabel("Petal Length")
plt.title("Line Plot")
plt.show()
```

```python
plt.figure(figsize =(6,4)) #size
plt.plot(iris_data["Id"][:20], iris_data["PetalLengthCm"][:20], color ="g", marker="o")
plt.xlabel("Id")
plt.ylabel("Petal Length")
plt.title("Line Plot")
plt.show()
```

Scatter Plot

```python
plt.figure(figsize =(6,4)) #size
plt.scatter(iris_data["Id"], iris_data["PetalLengthCm"], color ="b")
plt.xlabel("Id")
plt.ylabel("Petal Length")
plt.title("Scatter Plot")
plt.show()
```

```python
plt.figure(figsize =(6,4)) #size
plt.scatter(iris_data["Id"], iris_data["SepalLengthCm"], color ="b")
plt.xlabel("Id")
plt.ylabel("Sepal Length")
plt.title("Scatter Plot")
plt.show()
```

Histogram

```python
plt.figure(figsize =(6,4)) #size
plt.hist(iris_data["SepalLengthCm"], color ="b")
plt.ylabel("Frequency")
plt.xlabel("Sepal Length")
plt.title("Histogram")
plt.show()
```

```python
plt.figure(figsize =(6,4)) #size
plt.hist(iris_data["SepalLengthCm"], bins = 20, color ="b", edgecolor = "black")
plt.ylabel("Frequency")
plt.xlabel("Sepal Length")
plt.title("Histogram")
plt.show()
```

Box Plot

```python
plt.figure(figsize =(6,4)) #size
plt.boxplot(iris_data["SepalLengthCm"])
plt.ylabel("Sepal Length")
plt.title("Box Plot")
plt.show()
```

```python
plt.figure(figsize =(6,4)) #size
plt.boxplot(iris_data["PetalLengthCm"])
plt.ylabel("Petal Length")
plt.title("Box Plot")
plt.show()
```

```python
plt.figure(figsize =(6,4)) #size
plt.boxplot(iris_data["SepalWidthCm"])
plt.ylabel("Sepal Width")
plt.title("Box Plot")
plt.show()
```

```python
plt.figure(figsize =(6,4)) #size
plt.boxplot(iris_data["PetalWidthCm"])
plt.ylabel("Petal Width")
plt.title("Box Plot")
plt.show()
```

Numerical Data

```python
sns.distplot(tips_data["total_bill"])
```

Joint Plot

```python
sns.jointplot(x = tips_data["tip"],y =tips_data["total_bill"])
```

```python
plt.figure(figsize=(5,3))
sns.jointplot(x = tips_data["tip"],y =tips_data["total_bill"],kind="hex")
```

```python
plt.figure(figsize=(5,3))
sns.jointplot(x = tips_data["tip"],y =tips_data["total_bill"],kind="reg")
```
