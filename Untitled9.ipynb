```python
import pandas as pd
from scipy import stats

# Sample dataset (Replace with actual student marks)
marks = [85, 90, 78, 92, 85, 88, 76, 95, 89, 85]
# Convert to a Pandas Series
marks_series = pd.Series(marks)

# Compute Mean
mean_value = marks_series.mean()

# Compute Median
median_value = marks_series.median()

# Compute Mode
mode_value = stats.mode(marks_series)[0]

# Display results
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_value}")

```

```python
import numpy as np
# Sample dataset: Employee Salaries (in $)
salaries = [40000, 45000, 50000, 55000, 60000, 75000, 80000]
# Compute Range
range_value = max(salaries) - min(salaries)
# Compute Variance (Sample Variance: ddof=1)
variance_value = np.var(salaries, ddof=1)
# Compute Standard Deviation (Sample Standard Deviation: ddof=1)
std_dev_value = np.std(salaries, ddof=1)
# Display Results
print(f"Range: {range_value}")
print(f"Variance: {variance_value:.2f}")
print(f"Standard Deviation: {std_dev_value:.2f}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load Titanic dataset
df = sns.load_dataset("titanic")
# Display first 5 rows
print(df.head())
# Display dataset information
print("\nDataset Information:")
print(df.info())
# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())
# Handle missing values (Fill age with median, drop others)
df['age'].fillna(df['age'].median(), inplace=True)
df.drop(columns=['deck', 'embark_town', 'alive'], inplace=True)
# Univariate Analysis - Countplot of survival
plt.figure(figsize=(6, 4))
sns.countplot(x='survived', data=df, palette='coolwarm')
plt.title("Survival Count")
plt.show()
# Bivariate Analysis - Survival by class
plt.figure(figsize=(6, 4))
sns.barplot(x='class', y='survived', data=df, palette='coolwarm', estimator=lambda x: sum(x) /
len(x))
plt.title("Survival Rate by Class")
plt.show()
# Pairplot (Relationships between numerical variables)
sns.pairplot(df.select_dtypes(include=['number']), hue='survived', palette='coolwarm')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
# Create a sample dataset with missing values
data = {
'Age': [25, 30, np.nan, 35, 40, np.nan, 50, 55],
'Salary': [50000, 54000, 58000, np.nan, 62000, 66000, np.nan, 72000],
'City': ['New York', 'Los Angeles', np.nan, 'Chicago', 'Houston', 'Phoenix', 'Philadelphia',
np.nan]
}
df = pd.DataFrame(data)
# Display dataset with missing values
print("Original Dataset with Missing Values:\n", df)
# Mean Imputation for 'Age' and 'Salary'
mean_imputer = SimpleImputer(strategy='mean')
df[['Age', 'Salary']] = mean_imputer.fit_transform(df[['Age', 'Salary']])
# Median Imputation for 'Age' (alternative method)
median_imputer = SimpleImputer(strategy='median')
df['Age'] = median_imputer.fit_transform(df[['Age']]) # Replace age with median values
# Mode Imputation for 'City' (categorical column)
mode_imputer = SimpleImputer(strategy='most_frequent')
df[['City']] = mode_imputer.fit_transform(df[['City']])
# Display dataset after imputation
print("\nDataset After Imputation:\n", df)
```
