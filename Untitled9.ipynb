```python
import pandas as pd
from scipy import stats

# Sample dataset (Replace with actual student marks)
marks = [85, 90, 78, 92, 85, 88, 76, 95, 89, 85]
# Convert to a Pandas Series
marks_series = pd.Series(marks)

# Compute Mean
mean_value = marks_series.mean()

# Compute Median
median_value = marks_series.median()

# Compute Mode
mode_value = stats.mode(marks_series)[0]

# Display results
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_value}")

```

```python
import numpy as np
# Sample dataset: Employee Salaries (in $)
salaries = [40000, 45000, 50000, 55000, 60000, 75000, 80000]
# Compute Range
range_value = max(salaries) - min(salaries)
# Compute Variance (Sample Variance: ddof=1)
variance_value = np.var(salaries, ddof=1)
# Compute Standard Deviation (Sample Standard Deviation: ddof=1)
std_dev_value = np.std(salaries, ddof=1)
# Display Results
print(f"Range: {range_value}")
print(f"Variance: {variance_value:.2f}")
print(f"Standard Deviation: {std_dev_value:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# Sample dataset: Heights of males and females
data = pd.DataFrame({
"Gender": ["Male"] * 10 + ["Female"] * 10,
"Height": [170, 172, 175, 180, 185, 178, 182, 176, 185, 190, # Male Heights
158, 160, 162, 165, 168, 164, 166, 163, 169, 171] # Female Heights
})
# Create a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x="Gender", y="Height", data=data, palette="coolwarm")
# Title and labels
plt.title("Comparison of Heights: Males vs. Females", fontsize=14)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Height (cm)", fontsize=12)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load dataset (Example: Randomly generated sales data)
data = {
'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
'Sales_A': np.random.randint(100, 500, 12),
'Sales_B': np.random.randint(200, 600, 12),
'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South',
'East', 'West']
}
df = pd.DataFrame(data)
# Line Plot
plt.figure(figsize=(10,5))
plt.plot(df['Month'], df['Sales_A'], marker='o', linestyle='-', label='Sales A')
plt.plot(df['Month'], df['Sales_B'], marker='s', linestyle='--', label='Sales B')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Data')
plt.legend()
plt.grid(True)
plt.show()
```
