```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/Online Retail Data Set.xlsx")
df
```

**Boxplot to Visualize Outliers**

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df)
plt.title("Boxplot BEFORE Outlier Treatment")
plt.show()
```

**Detect Outliers Using IQR Rule**

```python
col = "Quantity"

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

Q1, Q3, IQR, lower_bound, upper_bound
```

**Identify Outliers**

```python
outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
outliers
```

**Handle Outliers**

**Remove Outliers**

```python
df_cleaned = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

**Cap Outliers (Winsorization)**
**Upper/lower extremes replaced with boundaries:**

```python
df_cap = df.copy()
df_cap[col] = df_cap[col].clip(lower=lower_bound, upper=upper_bound)
```

**Handle Missing Values (Median Imputation)**

```python
df[col] = df[col].fillna(df[col].median())
```

**Boxplot AFTER Outlier Treatment**

**For Removed Outliers:**

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_cleaned[col])
plt.title("Boxplot AFTER Removing Outliers")
plt.show()
```

**For Capped Outliers:**

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_cap[col])
plt.title("Boxplot AFTER Capping Outliers")
plt.show()
```

**Verify Results**

```python
df_cleaned.describe()
df_cap.describe()
```

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/employees.csv")
df
```

**Define Data Cleaning Function**

**Remove duplicates**

**Handle missing values**

**Allow choice of strategy: "mean", "median", "drop"**

```python
def clean_data(df, missing_strategy="mean"):
    cleaned_df = df.copy()

    print("Removing duplicates...")
    cleaned_df = cleaned_df.drop_duplicates()

    print("Handling missing values using:", missing_strategy)

    if missing_strategy == "mean":
        cleaned_df = cleaned_df.fillna(cleaned_df.mean(numeric_only=True))

    elif missing_strategy == "median":
        cleaned_df = cleaned_df.fillna(cleaned_df.median(numeric_only=True))

    elif missing_strategy == "drop":
        cleaned_df = cleaned_df.dropna()

    else:
        print("Invalid strategy! Using mean by default.")
        cleaned_df = cleaned_df.fillna(cleaned_df.mean(numeric_only=True))

    return cleaned_df
```

**Apply Function on Dataset**

```python
cleaned_df = clean_data(df, missing_strategy="median")
cleaned_df
```

**Check duplicates removed**

```python
print("Duplicates after cleaning:", cleaned_df.duplicated().sum())
```

**Check missing values handled**

```python
cleaned_df.isnull().sum()
```

**Compare shape before & after**

```python
print("Original shape:", df.shape)
print("Cleaned shape:", cleaned_df.shape)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Sample dataset
data = {
    'Student': ['A', 'B', 'C', 'D'],
    'Math': [45, 78, 62, 90],
    'Science': [50, 88, 72, 95]
}

df = pd.DataFrame(data)
print("Original Dataset:\n", df)

# MIN-MAX NORMALIZATION
minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[['Math', 'Science']] = minmax.fit_transform(df[['Math', 'Science']])

print("\nAfter Min-Max Normalization:\n", df_minmax)

# Z-SCORE STANDARDIZATION
zscore = StandardScaler()
df_zscore = df.copy()
df_zscore[['Math', 'Science']] = zscore.fit_transform(df[['Math', 'Science']])

print("\nAfter Z-score Standardization:\n", df_zscore)
```

```python
import pandas as pd

data = {
    'Student': ['A', 'B', 'C', 'C', 'D', None],
    'Marks': ['85', '90', None, '90', '70', '80'],
    'Date': ['2024-01-01', '2024-01-05', '2024-01-05', '2024-01-05', '2024-02-01', 'wrong']
}

df = pd.DataFrame(data)
print("Original Messy Dataset:\n", df)

# Identify Missing Values
print("\nMissing Values:\n", df.isnull().sum())


# Handle Missing Values
# Fill missing marks with mean
df['Marks'] = pd.to_numeric(df['Marks'], errors='coerce')  # convert to numeric
df['Marks'].fillna(df['Marks'].mean(), inplace=True)

# Fill missing student names with 'Unknown'
df['Student'].fillna('Unknown', inplace=True)


# Detect & Remove Duplicates
df = df.drop_duplicates()

# Fix Wrong Data Types
# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Verify Clean Dataset
print("\nCleaned Dataset:\n", df)
print("\nData Types After Cleaning:\n", df.dtypes)

# Save Clean Dataset
df.to_csv("cleaned_dataset.csv", index=False)
```

**Assignment-10**

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# IMPORT RAW DATASET
raw_data = {
    'Student': ['A', 'B', 'C', 'C', None, 'D'],
    'Marks': ['85', '90', None, '90', '70', '80'],
    'Age': [18, 19, 18, 18, None, 20],
    'Category': ['General', 'OBC', 'SC', 'SC', 'General', None]
}

df = pd.DataFrame(raw_data)
print("RAW DATASET:\n", df)

# CLEAN DATASET

# Fix datatypes
df['Marks'] = pd.to_numeric(df['Marks'], errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Handle missing values
df['Student'].fillna("Unknown", inplace=True)
df['Category'].fillna(df['Category'].mode()[0], inplace=True)
df['Marks'].fillna(df['Marks'].mean(), inplace=True)
df['Age'].fillna(df['Age'].median(), inplace=True)

# Remove duplicates
df = df.drop_duplicates()

# TRANSFORM DATASET

# Apply Min-Max Normalization on Marks and Age
scaler = MinMaxScaler()
df[['Marks_Scaled', 'Age_Scaled']] = scaler.fit_transform(df[['Marks', 'Age']])

# Encode categorical column
df['Category_Encoded'] = df['Category'].astype('category').cat.codes

# FINAL DATASET READY FOR ANALYSIS
print("\nFINAL CLEAN & TRANSFORMED DATASET:\n", df)

# Save dataset
df.to_csv("final_processed_dataset.csv", index=False)

# VALIDATION
print("\nData Types After Processing:\n", df.dtypes)
```
